```python
all_data =[]
for file in all_files:
    df = pd.read_excel(file)
    all_data.append(df)
all_data[0]
```

```python
all_data_concat = pd.concat(all_data, axis=0, ignore_index=True)
all_data_concat
```

```python
all_data_concat.to_csv('./data/riss_bigdata.csv', index=False, encoding='utf-8')
```

#### 제목

```python
all_title = all_data_concat['제목']
all_title
```

```python
all_data_concat = all_data_concat.drop (all_data_concat.columns[[0]], axis='columns')
all_data_concat
```

```python
stopwords = set(stopwords.words('english'))
stopwords
```

```python
lemma = WordNetLemmatizer()
```

```python
import re


words = []
for title in all_title:
    EnWords = re.sub(r'[^a-zA-Z]+',' ',str(title))
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopwords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    words.append(EnWordsTokenStopLemma)
```

```python
print(words) # 불용어 제거 후 남은 키워드 단어만 담기 
```

```python
words_reduce = list(reduce(lambda x, y: x+y, words)) # 2차원 리스트를 1차원 리스트로 축소하기 
print(words_reduce)
```

```python
count = Counter(words_reduce)
count
```

```python
word_count = dict()
for tag, counts in count. most_common(30):
    if len(str(tag))>1:
        word_count[tag] = counts
        print(f'{tag}:{counts}')
word_count
```

#### 단어개수 차트

```python
sorted_keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_values = sorted(word_count.values(),reverse=True)

plt.figure(figsize=(20,5))
plt.bar(range(len(word_count)),sorted_values,align='center')
plt.xticks(range(len(word_count)),list(sorted_keys),rotation=90)
plt.show()
```
